### Python Data Exploration

The pre-approved personal loan prediction dataset is designed to help predict the likelihood of a customer being approved for a personal loan. This dataset includes detailed information about customer demographics, financial details, and other relevant attributes that may influence the loan approval process.

### Column Name and Defination-------------------
ID:Unique identifier assigned to each individual.
Age: Age of the individual.
Education: Education level of the individual (e.g., Graduate, Undergraduate, Postgraduate).
Experience: Number of years of work experience.
Income:Monthly income of the individual.
ZIP Code: ZIP code of the individual's residence.
Married: Marital status of the individual (Yes/No).
Dependents:Number of dependents.
Credit_Card:Possession of a credit card (Yes/No).
Monthly_CC_Usage: Monthly credit card usage in dollars.
Credit_Score:Individual's credit score categories (Good, Standard and Poor)
Self_Employed:Self-employed status (Yes/No).
Salaried:Whether the individual salaried person. (Yes/No).
Loan_Amount: Loan amount applied for.
Nationality: Nationality of the personal or individual
Personal_Loan:Whether the individual has a personal loan (Yes/No).
Securities_Account:Possession of a securities account (Yes/No).
CD_Account:Possession of a certificate of deposit account (Yes/No).
Online_Banking:Online banking usage (Yes/No).
Last_Two_Years_Tax_Return_Filled:Filing of tax return (Yes/No).
Living_Property:Type of living property (Own/Rent).
Property_Area:Area of the property (Suburban/Urban/Rural).
Loan_Status:Loan approval status (Yes/No).

### Visual Exploratory Data Analysis(EDA)

Univariate Analysis: Exploring the data distribution of a SINGLE variable
Histogram : Exploring the distribution of a single Continuous(Quantitative) variable
Bar Charts : Exploring the distribution of a single Categorical variable

Bi-Variate Analysis: Exploring the relationship between TWO variables
Continuous Vs Continuous ---- Scatter Plot
Continuous Vs Categorical ---- Bar Plot, Box Plot
Categorical Vs Categorical ---- Cross Tabulation/Grouped Bar Plot

# Data Distribution: Histograms for continuous columns

## Continuous Vs Categorical Relationship ---- Bar Plot/Box Plot

```python
# Box plots for Categorical Target Variable "Loan_Status" and continuous predictors
ContiniousColsList=['Income', 'Age']

import matplotlib.pyplot as plt
fig, PlotCanvas=plt.subplots(nrows=1, ncols=len(ContiniousColsList), figsize=(10,10))

# Creating box plots for each continuous predictor against the Target Variable "Loan_Status"
for PredictorCol , i in zip(ContiniousColsList, range(len(ContiniousColsList))):
    LoanData.boxplot(column=PredictorCol, by='Loan_Status', figsize=(5,5), vert=True, ax=PlotCanvas[i])
```

```python
# Box Plot between Categorical and Continuous variable
%matplotlib inline
LoanData.boxplot(column='Monthly_CC_Usage', by='Loan_Status', figsize=(10,6), vert=True)
```

### Categorical Vs Categorical Relationship  ---- Grouped Bar Plot/Cross Tabulation

```python
# Cross tablulation between two categorical variables
CrossTabResult1=pd.crosstab(index=LoanData['Loan_Status'], columns=LoanData['Loan_Amount'])
CrossTabResult1
```

```python
# Visual Inference using Grouped Bar chart
CrossTabResult1.plot.bar(figsize=(10,3), color=['Red','blue','green','yellow'])
```

```python
# Visual Inference using Grouped Bar charts
CategoricalColsList=['Education', 'Married', 'Dependents','Credit_Card','Credit_Score','Self_Employed','Personal_Loan','Last_Two_Years_Tax_Return_Filled','Living_Property','Property_Area','CD_Account']

import matplotlib.pyplot as plt
fig, PlotCanvas=plt.subplots(nrows=len(CategoricalColsList), ncols=1, figsize=(10,100))

# Creating Grouped bar plots for each categorical predictor against the Target Variable "'Loan_Status'"
for CategoricalCol , i in zip(CategoricalColsList, range(len(CategoricalColsList))):
    CrossTabResult=pd.crosstab(index=LoanData[CategoricalCol], columns=LoanData['Loan_Status'])
    CrossTabResult.plot.bar(color=['red','green'], ax=PlotCanvas[i], title=CategoricalCol+' Vs '+'Loan_Status')
```
